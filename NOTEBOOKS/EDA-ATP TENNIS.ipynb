### TORNEO |Tournament |Series | Court | Surface

```python
dict_torneo = dict()
```

```python
def def_country(x):
    bandera = True
    for country in df_city_country['country'].unique():
            if country in x:
                bandera = False
                dict_torneo[x] = country  
    if bandera:
        for city in dict_city_country.keys():
            if city in x: 
                dict_torneo[x] = dict_city_country[city]
            else: 
                dict_torneo[x] = None
```

```python
for torneo in df['Tournament'].unique():
    def_country(torneo)

    
```

```python
df['Country'] = df['Tournament'].apply(lambda x: dict_torneo[x])
```
